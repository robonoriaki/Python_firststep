```python
#urlib module

import urllib.request

html = urllib.request.urlopen('http://google.co.jp/')
html_response = html.read()
print(html_response)
```

```python
url = 'https://www.google.co.jp/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png'

saveimage = 'googlelogo.png'

urllib.request.urlretrieve(url, saveimage)
print('saved')
```
