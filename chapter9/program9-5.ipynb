```python
#Draw a graph

import numpy as np
import matplotlib.pyplot as plt

#等差数列を作るarange()関数
#最小値、最大値、刻み幅を指定する
x = np.arange(0, 10, 0.1)
y = x
z = x - 1.0

#x軸とy軸を指定する
plt.plot(x, y, label = 'graph1')
plt.plot(x, z, label = 'graph2')

#凡例を作成する 凡例とは各グラフの名前が並んでいるところのこと
plt.legend()

#グラフを描画する
plt.show()
```

```python
y = np.sin(x)
plt.plot(x, y, label = 'graph1')
plt.legend()
plt.show()
```
